# Background and overview

Given the fact that college students, along with several other occupations like healthcare and security providers are suffering from insufficient sleep, this analysis aims at exploring the consequence of sleep deprivation. 

# Approach

We will calculate the correlation between sleeping time and several indicators of physical and mental health, as well as cognitive functions like memory. 

# Quick summary

Short sleep is positively correlated with high stress level, health issues like obesity, and worsened long-term memory. 



# Data

Briefly describe your dataset(s), including links to original sources.  Provide any relevant background information specific to your data sources.

```python
import numpy as np
import pandas as pd

# brief explanation of each dataset can be found below
depress = pd.read_csv('https://raw.githubusercontent.com/BohanMeng/storytelling-with-data/master/data-stories/sleep/sleep_stress.csv')
evaluation = pd.read_csv('https://raw.githubusercontent.com/BohanMeng/storytelling-with-data/master/data-stories/sleep/sleep_self_evaluation.csv')
ave=pd.read_excel("https://github.com/BohanMeng/storytelling-with-data/blob/master/data-stories/sleep/average_sleep_time.xlsx?raw=true")
memory = pd.read_csv("https://raw.githubusercontent.com/BohanMeng/storytelling-with-data/master/data-stories/sleep/sleep_memory.csv")
physical = pd.read_csv("https://raw.githubusercontent.com/BohanMeng/storytelling-with-data/master/data-stories/sleep/sleep_physical.csv")
```

# Analysis

Briefly describe each step of your analysis, followed by the code implementing that part of the analysis and/or producing the relevant figures.  (Copy this text block and the following code block as many times as are needed.)

```python
import plotly
import plotly.express as px
import plotly.graph_objects as go

# average sleeping time for people in the US is steadily increasing in the past years: contrary to my intuitions
fig3 = px.scatter(ave, x='Year', y='Estimate', trendline='ols', error_y="Standard Error", labels={'Estimate': 'estimated sleeping hour/day'}, title='sleep duration for people in the US')
fig3.show()

# feeling of tiredness negatively correlates with sleep time
fig2 = px.scatter(evaluation, x = 'Hours', y='Tired', trendline='ols', labels={'Hours':'sleep duration', "Tired": "self-evaluated level of tiredness"}, title='Self-evaluated level of tiredness with different sleeping hours')
fig2.show()

# sleep duration is negatively correlated with level of stress
fig = px.scatter(depress, x='sr.1', y='sl', trendline='ols', labels={'sr.1': 'sleep duration (hr)', 'sl': 'stress level'}, title='relations between stress level and sleep duration')
fig.show()
```

```python
# correlation between sleep duration and obesity is nonlinear
fig6 = px.scatter(physical, x='Sleep Time (hr)', y='Obese (%)', title='correlation between sleeping hours and obesity')
fig6.show()
```

```python
# correlation between sleep duration and illness is also nonlinear
# together with the last figure, we can tell that 7~9 hr/day is the most healthy sleeping time
fig7 = px.scatter(physical, x='Sleep Time (hr)', y='Existing illness/functional limitation (%)', title='correlation between sleeping hours and illness')
fig7.show()
```

```python
# disruption of sleep cause decreased performance in long-term memory task

fig5 = go.Figure()
fig5.add_trace(go.Histogram(x=memory['neuhitratediff_PBOPBO'], name='control'))
fig5.add_trace(go.Histogram(x=memory['neuhitratediff_DAMPPBO'], name='sleep disrupted'))

fig5.update_layout(title = 'memory performance after sleep-disrupting psychostimulants versus control', xaxis_title="difference in performance after sleep", yaxis_title="frequency",
    legend_title="group", barmode='overlay')
fig5.update_traces(opacity=0.75)
fig5.show()
```

# Interpretations and conclusions

Overall, people in the US have a lot of sleeping time in the recent decates, but this doesn't apply to all occupations. College students, healthcare providers, people in security related jobs, and several other groups are suffering from insufficient sleep. 

Lacking sleep could trigger in a number of consequences, including increased stress level and health issues like obisity. Meanwhile, sleep deprivation also reduce people's performance in long-term memory tasks, and this could lead to reduced work efficiency despite the fact that more time are spent on work.
