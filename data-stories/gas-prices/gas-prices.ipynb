# Data

Dataset provided by the US Energy Information Adminstration that includes weekly prices for various types of gas and diesel averaged over the country, region, select states, and select cities.

Link [here](https://www.eia.gov/dnav/pet/pet_pri_gnd_a_epm0_pte_dpgal_w.htm).

```python
xls = pd.ExcelFile('https://github.com/alexwells-22/storytelling-with-data/blob/master/data-stories/gas-prices/gas%20price%20data.xls?raw=true')
us_data = pd.read_excel(xls, 'Data 1')[2:].rename({'Back to Contents': 'Date', 'Data 1: Retail': 'Price'}, axis=1)
region_data = pd.read_excel(xls, 'Data 2')[2:].rename({'Back to Contents': 'Date', 'Data 2: Regions': 'East Coast', 'Unnamed: 2': 'New England', 'Unnamed: 3': 'Central Atlantic', 'Unnamed: 4': 'Lower Atlantic', 'Unnamed: 5': 'Midwest', 'Unnamed: 6': 'Gulf Coast', 'Unnamed: 7': 'Rocky Mountain', 'Unnamed: 8': 'West Coast', 'Unnamed: 9': 'West Coast except California'}, axis=1)

# Data available for future crunching, but not included here
# states_data = pd.read_excel(xls, 'Data 3')[2:]
# cities_data = pd.read_excel(xls, 'Data 4')[2:]

```

# Analysis

Briefly describe each step of your analysis, followed by the code implementing that part of the analysis and/or producing the relevant figures.  (Copy this text block and the following code block as many times as are needed.)

```python
fig = px.line(us_data, x='Date', y='Price')
fig.show()
```

```python
fig = px.line(region_data, x='Date', y='East Coast')
fig.show()
```

```python
fig = px.line(region_data, x='Date', y='East Coast')
fig.add_scatter(x=region_data['Date'], y=region_data['New England'], mode='lines')
fig.show()
```

```python
df = region_data.melt(id_vars=['Date']+list(region_data.keys()[10:]), var_name='Price')
fig = px.line(df, x='Date', y='value', color='Price')
fig.add_scatter(x=us_data['Date'], y=us_data['Price'], mode='lines', name = 'US Average')
```

```python
corr = pd.DataFrame
spearmanr(region_data['East Coast'], region_data['New England'])
corr.correlation
```
