# Background and overview

Winter is absolutely brutal in New England. Short days combined with cold weather make it difficult to get outside to get reasonable amounts of sunlight. We wondered if the amount of light people get might be related to their general happiness. The best sets of measures that we have are the [World Happiness Report](https://worldhappiness.report/ed/2021/) and data on [sunshine hours per city](https://en.m.wikipedia.org/wiki/List_of_cities_by_sunshine_duration). (check out this [super cool visualization](https://www.visualcapitalist.com/wp-content/uploads/2019/10/world-cities-by-hours-annual-sunshine.html) of the second data set).

# Approach

To figure out if there was a relationship between amount of sun and happiness, we correlated the annual total sunshine across various cities in each country we had data on with the happiness index of that country.

# Quick summary

Much to our surprise, we found a negative correlation between amount of sunlight a country recieves and its happiness index.

```python
#trend line for sunshine by country
fig = sunshine_by_country.drop('Year', axis=1).sort_values('Jul', ascending=False).T.plot(cmap='RdBu', legend=False)
fig
```

```python
sunshine_by_country1 = sunshine_by_country.copy()
happiness1 = happiness.copy()

#fips codes
fips_codes = pd.read_excel("https://github.com/ContextLab/storytelling-with-data/raw/master/data-stories/happy-sun/countryFIPS.xlsx").drop('ISO 3166-1', axis = 1)

#add fips code to data frame
def add_fips(df, fips_name = 'fips'):
  def lookup_fips(country):
    try:
      return fips_codes.query(f'`Country name` == "{country}"')['FIPS 10-4'].values[0]
    except IndexError:
      return np.nan

  df[fips_name] = df['Country'].apply(lookup_fips)
  return df

sunshine_by_country1 = add_fips(sunshine_by_country1.reset_index()).set_index('Country')
happiness1 = add_fips(happiness1.reset_index()).set_index('Country')
```

```python
happiness1.head()
```

```python
stacked_sunshine = pd.concat(12 * [pd.DataFrame(index=sunshine_by_country1.index, columns=['Month', 'Hours of sun', 'fips'])], axis=0)
stacked_sunshine['fips'] = pd.concat(12 * [pd.DataFrame(sunshine_by_country1['fips'])])
```

```python
for i, month in enumerate(sunshine_by_country.columns):
  if i >= 12:
    break

  start_ind = i * len(sunshine_by_country.index)
  end_ind = (i + 1) * len(sunshine_by_country.index)

  stacked_sunshine.iloc[start_ind:end_ind, 1] = sunshine_by_country1.iloc[:, i]
  stacked_sunshine.iloc[start_ind:end_ind, 0] = sunshine_by_country1.columns[i]
```

```python
stacked_sunshine['Hours of sun'] = stacked_sunshine['Hours of sun'].apply(float)
stacked_sunshine
```

# Interpretations and conclusions

We did not find the correlation that we anticipated between these measures. It seems that economic stability is correlated much more closely with happiness than the amount of sun that a particular country recieves in a year. Perhaps sunlamps for every global citizen wouldn't impact the average happiness of everyone. This data story could benefit from future exploration.
