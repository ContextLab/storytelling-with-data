# Data

```python
import numpy as np
import pandas as pd

# detailed info of each dataset can be find below
h19 = pd.read_csv('https://raw.githubusercontent.com/BohanMeng/storytelling-with-data/master/data-stories/happiness_report/report2019.csv')
h21 = pd.read_csv('https://raw.githubusercontent.com/BohanMeng/storytelling-with-data/master/data-stories/happiness_report/report2021.csv')
alcohol = pd.read_csv('https://raw.githubusercontent.com/BohanMeng/storytelling-with-data/master/data-stories/happiness_report/alcohol.csv')
```

# Analysis

Briefly describe each step of your analysis, followed by the code implementing that part of the analysis and/or producing the relevant figures.  (Copy this text block and the following code block as many times as are needed.)

```python
import plotly
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
```

```python
# economy, social support, health and freedom are strongly correlated with happiness
# generosity isn't strongly correlated with happiness
# surprisingly, perceptions of corruption is positively correlated with happiness -- both correlating with rising economy? 
c19 = h19.drop(["Overall rank"], axis=1).corr()
sns.heatmap(c19)
c19
```

```python
# surprisingly, the correlation between the health indicator and happiness score isn't greatly changed by COVID 
# economic, social support, health and freedom factors maintain the correlations
# generosity now negatively correlate with happiness
# perceptions of corruptions strongly negatively correlate with happiness: economic pressure caused by COVID makes it less tolerable? 
c21 = h21.filter(['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption'], axis=1).corr()
sns.heatmap(c21)
c21
```

```python
# strangely, GDP here is strongly negatively correlated with happiness, making me doubt the credibility of this dataset
# Besides that, alcohol consumption, human development index, and happiness are positively correlate with one another
# that doesn't suggest that drinking makes people happy; it could be economic factor instead 
sns.heatmap(alcohol.corr())
alcohol.corr()
```

# Interpretations and conclusions

Economic, social support, health and freedom are the key factors that positively correlate with happiness. COVID makes people less tolerable to corruption. Generosity doesn't play a significant role in determining happiness.
