```python
def lin_plot(x_name, y_name, x_title, y_title, dataset):
  sns.lmplot(x = x_name, y = y_name, fit_reg = False, hue = "Playoffs", data=dataset,legend = False)
  plt.xlabel(x_title)
  plt.ylabel(y_title)
  plt.legend(title='Playoffs', labels=['Not Made', 'Made'])
```

```python
lin_plot("W", "RS", "Wins", "Runs Scored", df)
lin_plot("W", "OBP", "Wins", "On Base Percentage", df)
lin_plot("W", "SLG", "Wins", "Slugging Percentage", df)
lin_plot("W", "BA", "Wins", "Batting Average", df)
```

```python
pre_2003  = df[df.Year <  2003]
post_2003 = df[df.Year >= 2003]

lin_plot("W", "RS",  "Wins", "Runs Scored", pre_2003)
lin_plot("W", "OBP", "Wins", "On Base Percentage", pre_2003)
lin_plot("W", "SLG", "Wins", "Slugging Percentage", pre_2003)
lin_plot("W", "BA",  "Wins", "Batting Average", pre_2003)
```

```python
lin_plot("W", "RS",  "Wins", "Runs Scored", post_2003)
lin_plot("W", "OBP", "Wins", "On Base Percentage", post_2003)
lin_plot("W", "SLG", "Wins", "Slugging Percentage", post_2003)
lin_plot("W", "BA",  "Wins", "Batting Average", post_2003)
```
