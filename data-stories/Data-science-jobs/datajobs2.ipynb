```python
# Provide code for downloading or importing your data here
url = "https://github.com/karimkhalil-byte/storytelling-with-data/raw/master/data-stories/Data-science-jobs/data_cleaned_2021.csv"
df = pd.read_csv(url).drop(labels=['Job Description','Salary Estimate','Company Name','Industry','Competitors','Hourly','Employer provided','Location','Python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch',
       'scikit', 'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo','google_an', 'job_title_sim', 'seniority_by_title', 'Degree','Headquarters'], axis=1).rename({'company_txt': 'Company Name'}, axis=1).set_index('index')
```

```python
df1 = df.query('`Job Title` == "Data Scientist"')
df2 = df1[df1.Founded != -1]
df2
```

```python
?px.bar
```

```python
df3 = df2.groupby("Founded").mean()
fig = px.bar(df3.reset_index(),x='Founded', y="Avg Salary(K)", )
fig.show()
```

```python
job_by_sector =  df2.groupby('Sector').mean()
job_by_revenue =  df2.groupby('Revenue').mean()
job_by_too =  df2.groupby('Type of ownership').mean()
job_by_size =  df2.groupby('Size').mean()
```

```python
job_by_sector1 = job_by_sector.sort_values('Avg Salary(K)', ascending=True)
```

```python
fig = px.bar(job_by_sector1.reset_index(), x= "Sector", y="Avg Salary(K)", color = "Rating")
fig.show()
```

```python
#fix x-axis, sort by x-axis
categories = ['Unknown / Non-Applicable',
              '$1 to $5 million (USD)',
              '$10 to $25 million (USD)',
              '$25 to $50 million (USD)', 
              '$50 to $100 million (USD)',
              '$100 to $500 million (USD)', 
              '$500 million to $1 billion (USD)',
              '$1 to $2 billion (USD)',
              '$2 to $5 billion (USD)',
              '$5 to $10 billion (USD)',              
              '$10+ billion (USD)']

job_by_revenue1 = job_by_revenue.copy().sort_values('Avg Salary(K)', ascending=True)
job_by_revenue1 = job_by_revenue1.rename({c: c.replace('$', '') for c in categories}, axis=0).reset_index()

fig = px.bar(job_by_revenue1, x = "Revenue", y="Avg Salary(K)", color = "Rating", 
             category_orders={'Revenue': [c.replace('$', '') for c in categories]})
fig.show()
```

```python
#fix x-axis, sort by x-axis

job_by_size1 = job_by_size.sort_values('Avg Salary(K)', ascending=True)
fig = px.bar(job_by_size1.reset_index(), x= "Size", y="Avg Salary(K)",
             color = "Rating",
             category_orders={'Size': ['1 - 50 ', '51 - 200 ', '201 - 500 ',
                                       '501 - 1000 ', '1001 - 5000 ',
                                       '5001 - 10000 ', '10000+ ']})
fig.show()
```

```python
#fix x-axis too
job_by_too1 = job_by_too.sort_values('Avg Salary(K)', ascending=True)
fig = px.bar(job_by_too1.reset_index(), x= "Type of ownership", y="Avg Salary(K)",color = "Rating")
fig.show()
```
