```python
import csv
import webbrowser as wb
import pandas as pd
import numpy as np
import collections

import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

from plotly import express as px
```

```python
liz_url = "https://github.com/saruulijile2/storytelling-with-data/raw/master/data-stories/Ed-nalysis/Liz.csv"
rose_url = "https://github.com/saruulijile2/storytelling-with-data/raw/master/data-stories/Ed-nalysis/Rose.csv"
ed_url = "https://github.com/saruulijile2/storytelling-with-data/raw/master/data-stories/Ed-nalysis/Ed-nalysis%20-%20Ed.csv"

liz = pd.read_csv(liz_url)
rose = pd.read_csv(rose_url)
ed = pd.read_csv(ed_url)

print(liz.head())
print(rose.head())
print(ed.head())
```

```python
# Liz bar graph categories
  
df = pd.DataFrame(liz)
  
categories = list(df["Category"])

for string in categories:
  if type(string) != str:
    categories.remove(string)

frequencies = {}
for item in categories:
    if item in frequencies:
        frequencies[item] += 1
    else:
        frequencies[item] = 1

# Sorting x axis alphabetically 
ordered_frequencies = collections.OrderedDict(sorted(frequencies.items()))

X = list(ordered_frequencies.keys())
Y = list(ordered_frequencies.values())

# Increase the size
plt.figure(figsize=(15,8))
plt.xticks(rotation=0)


# Plot the data using bar() method
plt.bar(X, Y, color='gold')
plt.title("Categories of Liz Tweets")
plt.xlabel("Tweet categories")
plt.ylabel("Number of Tweets")
  
# Show the plot
plt.show()
```

```python
# Liz bar graph valence
  
df = pd.DataFrame(liz)
  
categories = list(df["Valence"])

for string in categories:
  if type(string) != str:
    categories.remove(string)

frequencies = {}

neg = 0
neu = 0
pos = 0

for item in categories:
    if item == "Negative":
     neg = neg + 1
    if item == "Neutral":
      neu = neu + 1
    if item == "Positive":
      pos = pos + 1


frequencies['Negative'] = neg
frequencies['Neutral'] = neu
frequencies['Positive'] = pos

X = list(frequencies.keys())
Y = list(frequencies.values())

# Increase the size
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)

# Plot the data using bar() method
plt.bar(X, Y, color='gold')
plt.title("Valence of Liz Tweets")
plt.xlabel("Tweet valence")
plt.ylabel("Number of Tweets")
  
# Show the plot
plt.show()
print('negative: ', neg)
print('neutral: ', neu)
print('positive: ', pos)
```

```python
# analyzing the valence of Liz tweets using sentiment analyzer

import nltk
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
liz["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in liz["Comment"]]
liz["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in liz["Comment"]]
liz["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in liz["Comment"]]
liz_analyzer = liz[["Comment", "Positive", "Negative", "Neutral"]]
print(liz_analyzer.head())
```

```python
x = sum(liz_analyzer["Positive"])
y = sum(liz_analyzer["Negative"])
z = sum(liz_analyzer["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive")
    elif (b>a) and (b>c):
        print("Negative")
    else:
        print("Neutral")
sentiment_score(x, y, z)
print("Positive: ", x)
print("Negative: ", y)
print("Neutral: ", z)
```

```python
# Liz word cloud
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))

# Additional words to remove
stopword.add('liz')
stopword.add('day')
stopword.add('fianc')
stopword.add('big')
stopword.add('ed')
stopword.add('watch')
stopword.add('im')
stopword.add('singl')
stopword.add('life')

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
liz["Comment"] = liz["Comment"].apply(clean)
```

```python
text = " ".join(i for i in liz.Comment)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="black").generate(text)
plt.figure( figsize=(20,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
