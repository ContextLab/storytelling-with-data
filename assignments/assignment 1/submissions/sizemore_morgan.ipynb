Introduction


I tried to define a function that depends on two variables, x and y. There are three main arguments/instructions in the function: if, elif, and else. Each of these provides a different equation that depends on the relationship between x and y. If x>y, then a=x-y. If x<y, then a=x+y. Finally, if x is not greater than y and x is not less than y (x=y), then a=x*y. There are four possible different returns that are determined by which argument the user passes. If x>y and a is equal to 1,2,3, or 4 the function should return the words ‘little difference’, otherwise the function should return the value of a. If x<y, the function returns a. If x and y are equal and a is greater than nine, the function returns some value of x-y. If a is less than nine, the function returns a. In this function, all numerical values should be supported. 


```python
import numpy
dir(numpy)
```

```python
def func(x,y):
    if x>y:
        a=x-y
        return a    
    elif x<y:
        a=x+y
        return a
    else:
        a=x*y
        while a>9:
            return x-y
        else:
            return a
```

```python
print(func(3,2))
print(func(7,2))
print(func(2,3))
print(func(4,4))
print(func(2,2))
```
